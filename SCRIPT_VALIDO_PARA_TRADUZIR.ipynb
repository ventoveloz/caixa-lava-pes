```python
# Instala o tradutor
!pip install -q deep-translator

# Tradução da transcrição do Whisper (com chunk)
from deep_translator import GoogleTranslator
from google.colab import files

# Lê o conteúdo da transcrição
with open("/content/Conversa com aline.txt", "r", encoding="utf-8") as f:
    texto_original = f.read()

# Tamanho seguro por chunk
CHUNK_SIZE = 4500

# Divide em pedaços
texto_chunks = [texto_original[i:i + CHUNK_SIZE] for i in range(0, len(texto_original), CHUNK_SIZE)]

# Traduz e acumula os pedaços
traduzido_chunks = []
for chunk in texto_chunks:
    traduzido_chunks.append(GoogleTranslator(source='auto', target='pt').translate(chunk))

# Junta tudo
traduzido = "".join(traduzido_chunks)

# Salva a tradução
arquivo_traduzido = "/content/Conversa_com_aline_traduzido.txt"
with open(arquivo_traduzido, "w", encoding="utf-8") as f:
    f.write(traduzido)

# Baixa o arquivo
files.download(arquivo_traduzido)
```
